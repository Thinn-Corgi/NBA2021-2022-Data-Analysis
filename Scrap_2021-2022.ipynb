```python
import requests
import pandas as pd


```

```python

team_name = [
    "boston-celtics",
    "brooklyn-nets",
    "new-york-knicks",
    "philadelphia-76ers",
    "toronto-raptors",
    "chicago-bulls",
    "cleveland-cavaliers",
    "detroit-pistons",
    "indiana-pacers",
    "milwaukee-bucks",
    "atlanta-hawks",
    "charlotte-hornets",
    "miami-heat",
    "orlando-magic",
    "washington-wizards",
    "denver-nuggets",
    "minnesota-timberwolves",
    "oklahoma-city-thunder",
    "portland-trail-blazers",
    "utah-jazz",
    "golden-state-warriors",
    "los-angeles-lakers",
    "los-angeles-clippers",
    "phoenix-suns",
    "sacramento-kings",
    "dallas-mavericks",
    "houston-rockets",
    "memphis-grizzlies",
    "new-orleans-pelicans",
    "san-antonio-spurs"
]

def win(row):
    if'W' in row['Result']:
        return  'W'
    else: 
        return  'L'

    
def modify(df):
    df.fillna('0',inplace = True)
    df['Win'] = df.apply(lambda row: win(row), axis=1)# create a streak column
    df['Streak'] = df['Win'].groupby((df['Win'] != df['Win'].shift()).cumsum()).cumcount() + 1
    df['Streak'] =  df['Win'].astype(str) + df['Streak'].astype(str)
    df['Opponent'] = df['Opponent'].astype(str)
    df['Result'] = df['Result'].astype(str)
    df['FGM'] = df['FGM'].astype(int)
    df['FGA'] = df['FGA'].astype(int)
    df['FG%'] = df['FG%'].astype(float)
    df['FTM'] = df['FTM'].astype(int)
    df['FTA'] = df['FTA'].astype(int)
    df['FT%'] = df['FT%'].astype(float)
    df['3FGM'] = df['3FGM'].astype(int)
    df['3FGA'] = df['3FGA'].astype(int)
    df['3FG%'] = df['3FG%'].astype(float)
    df['PTS'] = df['PTS'].astype(int)
    df = df.rename(columns={'FG%' : "FG_PCT", 'FT%' : "FT_PCT", '3FG%' : "3FG_PCT" ,
                            'Date' : "DATE" , 'Opponent' : "OPP" , 'Result' : "RES" , 'FGM' : "FGM" ,
                            'FGA' : "FGA" , 'FTM' : "FTM" , 'FTA' : "FTA" ,
                            '3FGM' : "3FGM", '3FGA' : "3FGA" , 'PTS' : "PTS" , 'Win' : "WIN", 'Streak' : "STREAK"})





for i in range(0,len(team_name)):
    url = "https://www.foxsports.com/nba/{}-team-game-log".format(team_name[i])
    result = requests.get(url).text
    dfs = pd.read_html(result)
    df1 = dfs[0]
    modify(df1)
    filename = "{}2021_2022.csv".format(team_name[i])
    df1.to_csv(filename, index=False)

 
    

    
```
